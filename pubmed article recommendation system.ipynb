# **AST module**

The ast module provides a way to work with Abstract Syntax Trees (ASTs) in Python. An AST is a tree-like data structure that represents the structure of a piece of code as a set of nodes, each representing a different type of construct in the code, such as functions, classes, expressions, and so on. ASTs are used by the Python interpreter to parse and compile code, and can be useful for analyzing and manipulating code programmatically

```python
#it's help to convert string into list
import ast     #Abstract Syntax Trees (ASTs)
```

```python
pubmed.head(2)
```

```python
pubmed['abstractText']=pubmed['abstractText'].apply(lambda x:x.split())    #convert string into list and split element in the column
```

```python
pubmed.head(2)
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i)
    return L
```

```python
pubmed['meshMajor']=pubmed['meshMajor'].apply(convert)
```

```python
#it's remove the space betweem words
pubmed['meshMajor']=pubmed['meshMajor'].apply( lambda x:[i.replace (" ","")for i in x])
```

```python
pubmed['meshroot']=pubmed['meshroot'].apply(lambda x:x.split())
```

```python
import nltk
from nltk.stem import PorterStemmer

# Instantiate a Porter stemmer object
ps = PorterStemmer()

def conv(text):
    I=[]
    for i in text.split():
        I.append(ps.stem(i))
    return " ".join(I)
```

```python
df['tags']=df['tags'].apply(conv)
```

```python
cv.get_feature_names_out()
```

```python
def recommend(pubmed):
    pubmed_index=df[df['Title']==pubmed].index[0]
    distance=similarity[pubmed_index]
    pubmed_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in pubmed_list:
        print(df.iloc[i[0]].Title)
```
